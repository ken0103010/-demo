### (2)上網搜尋scipy.signal.freqz，使用此函式同樣繪製bk={1, 2, 1}之FIR filter之頻率響應。這個函式繪製出來的頻率響應一樣嗎？

## (1)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

w_hat = np.arange(-np.pi, np.pi, np.pi/500)
H = 1 + 2*np.exp(-1j * w_hat) + np.exp(-1j * 2 * w_hat)

plt.figure(figsize=[15,5])

plt.subplot(121)
plt.plot(w_hat, abs(H))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The magnitude spectrum')

plt.subplot(122)
plt.plot(w_hat, np.angle(H/abs(H)))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The phase spectrum')
```

## (2)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

w_hat = np.arange(-np.pi, np.pi, np.pi/50)
bk = [1,2,1]

w, h = signal.freqz(b=bk, a=1, worN = w_hat)


plt.figure(figsize=[15,5])

plt.subplot(121)
plt.plot(w, abs(h))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The magnitude spectrum')

plt.subplot(122)
plt.plot(w, np.angle(h/abs(h)))
plt.xlabel(r'$\hat\omega$  (radian)')
plt.title('The phase spectrum')
```

## (1)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

b = [1, -1]
a = [1]

p = np.roots(a)
z = np.roots(b)

angle = np.linspace(-np.pi, np.pi, 50)
cirx = np.sin(angle)
ciry = np.cos(angle)
plt.figure(figsize=(8,8))
plt.plot(cirx, ciry,'k-')
plt.plot(np.real(z), np.imag(z), 'o', markersize=12)
plt.plot(np.real(p), np.imag(p), 'x', markersize=12)
plt.grid()

plt.xlim((-2, 2))
plt.xlabel('Real')
plt.ylim((-2, 2))
plt.ylabel('Imag')
```
